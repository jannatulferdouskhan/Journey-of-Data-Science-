```python
import pandas as pd 
from matplotlib import pyplot as plt 
df = pd.read_csv("nasdaq100.csv",sep=';') 
df 
df= df.drop(columns=['Date']) #Columns delete  
df.head() 


```

```python
plt.scatter(df["Starting (USD)"] ,df["Ending (USD)"])
plt.xlabel("Starting (USD)")
plt.ylabel("Ending (USD)")
plt.title("Nasdaq 100 Stock Prices")
```

```python
x= df.drop('Ending (USD)',axis=1)
x.head()
```

```python
y=df[["Ending (USD)"]]
y.head()
```

```python
#Linear Regrassion 
from sklearn.linear_model import LinearRegression
reg = LinearRegression() 
x.mean()
```

```python
y.mean()
```

```python
plt.scatter(x.mean(),y.mean(), color='red')
plt.scatter(df["Starting (USD)"] ,df["Ending (USD)"])
plt.xlabel("Starting (USD)")
plt.ylabel("Ending (USD)")
plt.title("Nasdaq 100 Stock Prices")


```

```python
reg.fit(x,y)
```

```python
# #Manual 
m=reg.coef_
m
c=reg.intercept_
c
# y= mx + c

m*16700 + c

# Predict USing function 
reg.predict([[16700]])
```

```python
df['Predected_Y'] = reg.predict(x)
df
```

```python
plt.plot(x,reg.predict(x))
plt.scatter(x.mean(),y.mean(), color='red')
plt.scatter(df["Starting (USD)"] ,df["Ending (USD)"])
plt.xlabel("Starting (USD)")
plt.ylabel("Ending (USD)")
plt.title("Nasdaq 100 Stock Prices")
```

```python
# Loss and cost Function 
df['Lose']= df['Ending (USD)'] - df['Predected_Y']
df.head() 

```

```python
# MSE
from sklearn.metrics import mean_absolute_error, mean_squared_error 
mse= mean_squared_error(df['Ending (USD)'],df['Predected_Y'])  
mse
```

```python
# MAE
from sklearn.metrics import mean_absolute_error, mean_squared_error 
mae= mean_absolute_error(df['Ending (USD)'],df['Predected_Y'])  
mae
```

```python
#MAE manualy 
sum(abs(df['Lose'])) / len(x)
```

```python
#Model Performence 
reg.score(x,y)


```

# Salary Prediction using Linear Regression

```python
# Read File 
import pandas as pd 
from matplotlib import pyplot as plt 
df= pd.read_csv("Salary Data.csv",sep=",")
df.head()
```

# EDA

```python
# Basic Info Check 
df.info()
df.describe()
df.isnull().sum()
```

```python
# Univariate Analysis (Single Column)
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df['Experience Years'],kde=True)
plt.show()
# Salary 
sns.histplot(df['Salary'], kde=True)
plt.show()
```

```python
# Bivariate Analysis (Relationship between 2 columns)
sns.scatterplot(x="Experience Years",y="Salary",data=df)
plt.show()
```

```python
# Correlation Check 
df.corr()
sns.heatmap(df.corr(), annot=True , cmap='coolwarm')
plt.show()
```

```python
# Boxplot for Outliers 
sns.boxenplot(x=df['Salary'])
plt.show()
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression 

reg=LinearRegression()
x = df.drop("Salary",axis=1)
y = df[["Salary"]]
x.head()
y.head()



plt.scatter(x.mean(),y.mean(),color="Red")

plt.scatter(df["Experience Years"],df["Salary"])
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.title("Exprience $ Salary")
plt.show() 

reg.fit(x,y)

```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error 
df['Predected_Salary'] = reg.predict(x)
df.head() 

 
```

```python
plt.plot(x,reg.predict(x))
plt.scatter(df["Experience Years"],df["Salary"])
plt.scatter(df["Experience Years"], df['Predected_Salary'] ,color="yellow")
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.title("Exprience $ Salary")
plt.show() 
reg.score(x,y)
```
